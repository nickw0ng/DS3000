```python
import random

def magic_8_ball(question, shakes):
 
    possible_answers = [
        "Yes",
        "Maybe not ",
        "Not sure",
        "I'm busy right now, try again in 5 minutes",
        "Not the best idea",
        "No",
        "100%",
        "0%"
    ]
    
    ball_answer = random.choices(possible_answers)
 
    print(f"{question}")
    print(f"After {shakes} shakes the magic 8 ball says: {ball_answer[-1]}")
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_crash = sns.load_dataset('car_crashes')
print(df_crash)
'''
print(df_crash.head(6))

df_selected_columns = df_crash[['total', 'ins_premium', 'ins_losses', 'abbrev']]

print(df_selected_columns) 
df_ins_premium = df_crash[df_crash['ins_premium'] >= 1000]

print(df_ins_premium)

average_speed = df_crash['speeding'].mean()

df_speed = df_crash[df_crash['speeding'] > average_speed]

print(df_speed)


numerical_columns = df_crash.drop(columns=['abbrev'])
                                           
medians = numerical_columns.median()

print(medians)

correlation = df_crash['alcohol'].corr(df_crash['ins_premium'])


print(f"Correlation between alcohol and ins_premium: {correlation}")
print('Because the correlation coefficient is -0.1706, means there is a weak negative correlation. In other words, the dataset suggets a weak link of between alcohol and insurance premium so another factor could  ') '''

west = ['AK', 'HI', 'CA', 'OR', 'WA', 'ID', 'NV', 'UT', 'CO', 'WY', 'MT']
southwest = ['AZ', 'NM', 'TX', 'OK']
midwest = ['ND', 'SD', 'NE', 'KS', 'MN', 'IA', 'MO', 'WI', 'IL', 'IN', 'MI', 'OH']
southeast = ['AR', 'LA', 'MS', 'TN', 'KY', 'AL', 'GA', 'FL', 'SC', 'NC', 'VA', 'WV', 'DE', 'MD', 'DC']
northeast = ['PA', 'NJ', 'CT', 'RI', 'NY', 'MA', 'NH', 'VT', 'ME']

region = list()
for st in range(51):
    if df_crash.abbrev[st] in west:
        region.append('west')
    elif df_crash.abbrev[st] in southwest:
        region.append('southwest')
    elif df_crash.abbrev[st] in midwest:
        region.append('midwest')
    elif df_crash.abbrev[st] in southeast:
        region.append('southeast')
    else:
        region.append('northeast')

df_crash['region'] = region

colors = {
    'west': 'red',
    'southwest': 'orange',
    'midwest': 'yellow',
    'southeast': 'green',
    'northeast': 'blue'
}

plt.figure(figsize=(12, 6))
for region, color in colors.items():
    plt.scatter(df_crash[df_crash['region'] == region]['alcohol'],
                df_crash[df_crash['region'] == region]['ins_premium'],
                label=region, color=color, alpha=0.7, edgecolor='black')

plt.title('Alcohol vs Insurance Premium per Region')
plt.xlabel('alcohol')
plt.ylabel('ins_premium')
plt.legend()

plt.savefig('alcohol_vs_ins_premium.pdf')

plt.show()

```
