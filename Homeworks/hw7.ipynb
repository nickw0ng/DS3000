## Part 3.1 (10 points)
Build a 20-Nearest Neighbor classifier which is `fit` and then `predict` on the whole dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay


df_ev = pd.read_csv('evdataset.csv', index_col='id')
```

```python
# The needed data 
X = df_ev.drop(columns=['Drive'])
y = df_ev['Drive']

# Initializing the classifier and dataset
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X, y)
y_pred = knn.predict(X)

# Gets the matrix
conf_mat = confusion_matrix(y, y_pred, labels=knn.classes_)
```

## Part 3.2 (10 points)
Plot the confusion matrix which results from this dataset.  

Using the confusion matrix, identify the drive mostly commonly misclassified by
- explicitly computing its accuracy
- telling which other drive it is most commonly mistaken for

Your response should be one or two sentences.

```python
conf_mat_disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=knn.classes_)
conf_mat_disp.plot()
plt.grid(False)
```

```python
accuracy_per_class = conf_mat.diagonal() / conf_mat.sum(axis=1)

# Identify the most commonly misclassified drive
most_misclassified_index = np.argmin(accuracy_per_class)
most_misclassified_drive = knn.classes_[most_misclassified_index]
most_misclassified_accuracy = accuracy_per_class[most_misclassified_index]

# Identify which drive it is most commonly mistaken for
mistake_counts = conf_mat[most_misclassified_index]
mistake_counts[most_misclassified_index] = 0  
most_common_mistake_index = np.argmax(mistake_counts)
most_common_mistake_drive = knn.classes_[most_common_mistake_index]

print(f"Most misclassified drive: {most_misclassified_drive} with accuracy {most_misclassified_accuracy:.2f}")
print(f"Most commonly mistaken for: {most_common_mistake_drive}")
```
