## Memory Management

By default, each of the methods above returns a **copy** of the dataframe. Instead of manually overwriting or creating a new data frame:

If you want to replace the dataframe with the output of some method, set `inplace=True` (available on many methods)

**Note**: this has the benefit of saving memory; instead of creating duplicates of a large data set, you update the data set itself. *However* you must be careful when making the changes, or you may have to re-start everything to get a clean data set.

````python
df_poke = pd.read_csv('pokedata.csv')
df_poke = df_poke.rename(make_lower, axis = 1)
df_poke = df_poke.set_index('pokedex')
df_poke.head()
````

````python
df_poke.sort_index(axis = 1).head()
````

````python
df_poke.head()
````

````python
df_poke.sort_index(axis = 1, inplace = True)
df_poke.head()
````

Common mistakes with `inplace`:
- assuming that a method/function works `inplace` when it doesn't
    - effect: your object hasn't been manipulated (though you wish it had been)
- assuming that a method/function returns a copy of object when it operates `inplace`
    - effect: your original data has been modified
        - symptom: your return value is `None`
        - symptom: your data appears to be changing 'by itself'

## Working with Duplicates

* Sometimes datasets contain duplicates record, which you may want to discard

### duplicated() method

Returns a boolean Series indicating whether each row is a duplicated (whether it has been observed in a previous row)

````python
df_dupe = pd.read_excel('dupe_data.xlsx', na_values = ('miss'))
df_dupe
````

````python
# even though row 6 and 7 are identical, only 7 is marked
df_dupe.duplicated()
````

````python
df_dupe.duplicated(subset = 'patient')
````

### drop_duplicates() method

* Returns a DataFrame where the duplicates have been dropped

````python
df_dupe
````

````python
df_dupe.drop_duplicates()
````

* You can also specify a subset of columns to detect duplicates.
* Pass in a list of column names as an argument to the **drop_duplicates()** method call.

````python
df_dupe.drop_duplicates(subset = 'patient')
````

* By default, both duplicated() and drop_duplicates() keep the first observed value combination. 
* If you want to keep the last value, pass in **keep="last"**
* This can be used for example where data is ordered chronologically and the last mention is the most recent one (here: patient g had a duplicate record because their trial 2 data was eventually found)

````python
df_dupe.drop_duplicates(subset = 'patient', keep = 'last')
````
